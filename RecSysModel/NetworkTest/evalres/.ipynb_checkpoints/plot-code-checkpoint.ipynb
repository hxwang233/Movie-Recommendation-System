```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 收敛曲线

```python
x = range(0,205,5)
y_label = ['svd','GMF','NCF','convNCF','ENMF']
for y in y_label:
    hr = np.loadtxt("./"+str.lower(y)+"/hr_list_200epoch.txt")
    plt.plot(x, hr, label=y)
plt.xlim(0, 200)
plt.ylim(0, 0.5)
plt.xlabel('Epoch')
plt.ylabel('HR@100')
plt.legend(loc='upper right')
plt.savefig("../evalres/lab-picture/Epoch vs. HR@100.png")
```

```python
x = range(0,205,5)
y_label = ['SVD','GMF','NCF','convNCF','ENMF']
for y in y_label:
    ndcg = np.loadtxt("./"+str.lower(y)+"/ndcg_list_200epoch.txt")
    plt.plot(x, ndcg, label=y)
plt.xlim(0, 200)
plt.ylim(0, 0.2)
plt.xlabel('Epoch')
plt.ylabel('NDCG@100')
plt.legend(loc='upper right')
plt.savefig("../evalres/lab-picture/Epoch vs. NDCG@100.png")
```

##  negative weight

```python
x = [1, 2, 3, 4, 5, 6, 7]
y_label = ['ENMF']
fig, ax1 = plt.subplots(figsize = (8, 5))
ax1.set_xlabel('Negative weight')
ax1.set_ylabel('HR@100')
ax2 = ax1.twinx()
ax2.set_ylabel('NDCG@100')

plt.xticks(x,(0, 0.1, 0.3, 0.5, 0.7, 0.9, 1))
ax1.set_ylim(0, 0.55)
ax2.set_ylim(0, 0.2)

for y in y_label:
    hr = np.loadtxt("./"+str.lower(y)+"/hr_list_neg_weight.txt")
    ax1.plot(x, hr, label='ENMF-HR@100')
    ax1.scatter(x, hr)
for y in y_label:
    ndcg = np.loadtxt("./"+str.lower(y)+"/ndcg_list_neg_weight.txt")
    ax2.plot(x, ndcg, color='g',label='ENMF-NDCG@100')
    ax2.scatter(x, ndcg, color='g')
ax1.legend(loc='upper left')
ax2.legend(loc='center right')
plt.savefig("../evalres/lab-picture/Negative weight  vs. HR@100 and NDCG@100.png")
```

## topK

```python
y_label = ['UserCF','ItemCF','ItemPop','ItemPop-NCF','ItemPop-convNCF','ItemPop-ENMF','SVD','GMF','NCF','convNCF','ENMF']
data = {'\\': y_label, 'HR@50':list(), 'HR@100':list(), 'HR@200':list(), 'NDCG@50':list(), 'NDCG@100':list(), 'NDCG@200':list()}
for y in y_label:
    hr = np.loadtxt("./"+str.lower(y)+"/hr_list_topk.txt")
    for i in range(len(hr)):
        key = 'HR@'+str(50 * int(np.power(2,i)))
        data[key].append(round(hr[i], 4))
    ndcg = np.loadtxt("./"+str.lower(y)+"/ndcg_list_topk.txt")
    for i in range(len(ndcg)):
        key = 'NDCG@'+str(50 * int(np.power(2,i)))
        data[key].append(round(ndcg[i], 4)) 

df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(20, 20))
ax.axis('off')
ax.axis('tight')
tb = plt.table(cellText=df.values,colLabels=df.columns,bbox=[0, 0, 1, 1],cellLoc='center')
tb.set_fontsize('xx-large')
plt.savefig("../evalres/lab-picture/topk table.png")
```

## time

```python
y_label = ['UserCF','ItemCF','SVD','GMF','NCF','ConvNCF','ENMF']
data = {'\\': y_label, 'single(s)':list(), 'total(s)':list(), 'appr. conver epoch':["","",50,10,10,80,100]}
for i,y in enumerate(y_label):
    if y=='UserCF' or y=='ItemCF':
        time = np.loadtxt("./"+str.lower(y)+"/time.txt")
        data['single(s)'].append("")
        data['total(s)'].append(np.around(float(time),2))
    else:
        time = np.loadtxt("./"+str.lower(y)+"/single_time.txt")
        data['single(s)'].append(np.around(float(time),2))
        data['total(s)'].append(np.around(float(time)*data['appr. conver epoch'][i], 2))
df = pd.DataFrame(data)
fig, ax = plt.subplots(figsize=(20, 20))
ax.axis('off')
ax.axis('tight')
tb = plt.table(cellText=df.values,colLabels=df.columns,bbox=[0, 0, 1, 1],cellLoc='center')
tb.set_fontsize('xx-large')
plt.savefig("../evalres/lab-picture/time table.png")
```
